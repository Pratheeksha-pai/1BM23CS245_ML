```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes, fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Function to evaluate model
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print("MSE  :", mse)
    print("RMSE :", rmse)
    print("MAE  :", mae)
    print("R2   :", r2)
    print("-" * 40)

# Load Dataset 1: Diabetes
diabetes = load_diabetes()
X1 = diabetes.data
y1 = diabetes.target

# Train-test split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# CART Model
cart1 = DecisionTreeRegressor(criterion="squared_error", max_depth=4, random_state=42)
cart1.fit(X_train1, y_train1)

# Predictions
y_pred1 = cart1.predict(X_test1)

print("Results for Diabetes Dataset")
evaluate_model(y_test1, y_pred1)
```

```python
# Load Dataset 2: California Housing
housing = fetch_california_housing()
X2 = housing.data
y2 = housing.target

# Train-test split
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42)

# CART Model
cart2 = DecisionTreeRegressor(criterion="squared_error", max_depth=6, random_state=42)
cart2.fit(X_train2, y_train2)

# Predictions
y_pred2 = cart2.predict(X_test2)

print("Results for California Housing Dataset")
evaluate_model(y_test2, y_pred2)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12,6))
plot_tree(cart1, filled=True)
plt.title("CART Regression Tree - Diabetes")
plt.show()
```
